```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output2.csv")
bends = pd.read_csv("bends.csv")
```

```python
# Fexperimental vs Reynolds number (for pipe 3)
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="Re", y="Fexp",
             data=data.loc[data["D"]==9.6, ["Fexp", "Re"]], marker="o")
plt.xlabel("Reynolds Number (Re)")
plt.ylabel("Friction factor (exp)")
plt.title("Experimental Friction Factor v/s Reynolds Number (Pipe 3)")
```

```python
#F experimental and F theoritical vs Reynolds number for pipe 2 

sns.set_theme(style="darkgrid")
sns.lineplot(
    x="Re",
    y="Fexp",  
    marker="o",  # Markers for better visibility
    data=data.loc[data["D"]==7.8],
    label="F(experimental)"
)
plt.xlabel("Reynolds Number")
plt.ylabel("Friction Factor (f)")
plt.title("Friction Factor vs Reynolds Number (Pipe 2)")

sns.lineplot(
    x="Re",
    y="Ftheo",
    marker="X",  # Markers for better visibility
    data=data.loc[data["D"]==7.8],
    label="F(theoritical)"
)

x_err = data.loc[data["D"] == 7.8, "dRe/Re"] * data.loc[data["D"] == 7.8, "Re"] / 100.0
y_err = data.loc[data["D"] == 7.8, "dFe/Fe"] * data.loc[data["D"] == 7.8, "Fexp"] / 100.0
plt.errorbar(
    data.loc[data["D"] == 7.8, "Re"], 
    data.loc[data["D"] == 7.8, "Fexp"], 
    xerr=x_err, yerr=y_err, 
    fmt="none",  # No markers (already in lineplot)
    capsize=5, color="black", alpha=0.5 
)

plt.xlabel("Reynolds Number")
plt.ylabel("Friction Factor (f)")
plt.title("Friction Factor vs Reynolds Number (Pipe 2)")
plt.legend()
plt.show()
```
