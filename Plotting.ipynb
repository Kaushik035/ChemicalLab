```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output2.csv")
bends = pd.read_csv("bends.csv")
```

```python
# Fexperimental vs Reynolds number (for pipe 3)
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="Re", y="Fexp",
             data=data.loc[data["D"]==9.6, ["Fexp", "Re"]], marker="o")
plt.xlabel("Reynolds Number (Re)")
plt.ylabel("Friction factor (exp)")
plt.title("Experimental Friction Factor v/s Reynolds Number (Pipe 3)")
```

```python
#F experimental and F theoritical vs Reynolds number for pipe 2 

sns.set_theme(style="darkgrid")
sns.lineplot(
    x="Re",
    y="Fexp",  
    marker="o",  # Markers for better visibility
    data=data.loc[data["D"]==7.8],
    label="F(experimental)"
)
plt.xlabel("Reynolds Number")
plt.ylabel("Friction Factor (f)")
plt.title("Friction Factor vs Reynolds Number (Pipe 2)")

sns.lineplot(
    x="Re",
    y="Ftheo",
    marker="X",  # Markers for better visibility
    data=data.loc[data["D"]==7.8],
    label="F(theoritical)"
)
plt.xlabel("Reynolds Number")
plt.ylabel("Friction Factor (f)")
plt.title("Friction Factor vs Reynolds Number (Pipe 2)")
plt.legend()
plt.show()
```

```python
#F experimental and F theoritical vs Reynolds number for pipe 4

sns.set_theme(style="darkgrid")
sns.lineplot(
    x="Re",
    y="Fexp",  
    marker="o",  # Markers for better visibility
    data=data.loc[data["D"]==15.8],
    label="F(experimental)"
)
plt.xlabel("Reynolds Number")
plt.ylabel("Friction Factor (f)")
plt.title("Friction Factor vs Reynolds Number (Pipe 4)")

sns.lineplot(
    x="Re",
    y="Ftheo",
    marker="X",  # Markers for better visibility
    data=data.loc[data["D"]==15.8],
    label="F(theoritical)"
)
plt.xlabel("Reynolds Number")
plt.ylabel("Friction Factor (f)")
plt.title("Friction Factor vs Reynolds Number (Pipe 4)")
plt.legend()
plt.show()
```

```python
# dRe/Re vs dFe/Fe for Pipe 2 
ax = sns.barplot(
    x="dRe/Re",
    y="dFe/Fe",  
    palette="deep", 
    data=data.loc[data["D"] == 7.8]
)
ax.set_xticklabels([f'{val:.3f}' for val in ax.get_xticks()], rotation=45, ha='right')
plt.xlabel(r'%Error in Reynolds Number: $\bigtriangleup Re / Re$')
plt.ylabel(r'%Error in Friction Factor: $\bigtriangleup F / F$')
plt.title("%Error in Friction Factor v/s %Error in Reynolds Number (Pipe 2)")
plt.grid(visible=True, which='both', linestyle='--', linewidth=1)
plt.show()
```

```python
# dRe/Re vs dFe/Fe for Pipe 3
ax = sns.barplot(
    x="dRe/Re",
    y="dFe/Fe",  
    palette="deep", 
    data=data.loc[data["D"] == 9.6]
)
ax.set_xticklabels([f'{val:.3f}' for val in ax.get_xticks()], rotation=45, ha='right')
plt.xlabel(r'%Error in Reynolds Number: $\bigtriangleup Re / Re$')
plt.ylabel(r'%Error in Friction Factor: $\bigtriangleup F / F$')
plt.title("%Error in Friction Factor v/s %Error in Reynolds Number (Pipe 3)")
plt.grid(visible=True, which='both', linestyle='--', linewidth=1)
plt.show()
```

```python
# dRe/Re vs dFe/Fe for Pipe 4
ax = sns.barplot(
    x="dRe/Re",
    y="dFe/Fe",  
    palette="deep", 
    data=data.loc[data["D"] == 15.8]
)
ax.set_xticklabels([f'{val:.3f}' for val in ax.get_xticks()], rotation=45, ha='right')
plt.xlabel(r'%Error in Reynolds Number: $\bigtriangleup Re / Re$')
plt.ylabel(r'%Error in Friction Factor: $\bigtriangleup F / F$')
plt.title("%Error in Friction Factor v/s %Error in Reynolds Number (Pipe 4)")
plt.grid(visible=True, which='both', linestyle='--', linewidth=1)
plt.show()
```
